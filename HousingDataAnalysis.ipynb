```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Housing.csv")
```

```python
print("\n===================================== Studying the data =====================================\n")
print("Starting 5 lines:")
data.head()
```

```python
print("\n ********************** Observations taken from whole data: ********************** \n")
print(data.describe(include = "all"))
```

```python
print(data.count())
```

```python
print("\n***************************** Checking missing values *****************************\n")
print(data.isnull().sum())
data = data.dropna() #it will remove null or empty values
```

```python
targeted_columns = ["price", "area", "bedrooms"]
for col in targeted_columns:
  print(f"\n ******************** Data for {col} ******************** \n")
  print("Mean         : ", data[col].mean())
  print("Median       : ", data[col].median())
  print("Mode         : ", data[col].mode()[0])
  print("Variance     : ", data[col].var())
  print("Std dev      : ", data[col].std())
  print("Skewness     : ", data[col].skew())
  print("Kurtosis     : ", data[col].kurt())
  print("Max          : ", data[col].max())
  print("Min          : ", data[col].min())
  print("\n Note: If the difference btw Mean and Std dev is large then std dev is high")
```

Histogram for Price and Area  


```python
# Graph plotting for visualization (By matplot)
# Histogram for price vs area

data["price"].hist(bins = 30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
data["area"].hist(bins = 30)
plt.title("Area Distribution")
plt.xlabel("Area")
plt.ylabel("Frequency")
plt.show()
```
